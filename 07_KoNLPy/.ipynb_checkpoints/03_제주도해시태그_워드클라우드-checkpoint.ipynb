```python
import pandas as pd
```

```python
raw_total = pd.read_excel('./3_1_crawling_raw.xlsx')
raw_total
```

```python
tags_total = list()

for tags in raw_total['tags']:
    split_list = tags[2:-2].split("', '")
    for tag in split_list:
        tags_total.append(tag)
    
tags_total
```

```python
from collections import Counter
```

```python
tags_counts = Counter(tags_total)
tags_counts
```

## 가장 많은 count 조회

```python
tags_counts.most_common(50)
```

## 키워드 정제하기

```python
STOPWORDS = ['#일상', '#선팔', '#제주도', '#jeju', '#반영구', '#제주자연눈썹',
             '#서귀포눈썹문신', '#제주눈썹문신', '#소통', '#맞팔']
```

```python
tags_total_selected = list()

for tag in tags_total:
    if tag not in STOPWORDS:
        tags_total_selected.append(tag)
        
tags_total_selected        
```

```python
tags_total_selected = Counter(tags_total)
tags_total_selected
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
from matplotlib import font_manager, rc
import platform
```

```python
path = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname = path).get_name()
rc('font', family = font_name)
```

## 막대그래프

```python
tag_counts_df = pd.DataFrame(tags_total_selected.most_common(30))
tag_counts_df
```

```python
tag_counts_df.columns = ['tags', 'counts']
tag_counts_df
```

```python
plt.figure(figsize = (10, 8))
sns.barplot(x = 'counts', y = 'tags', data = tag_counts_df)
```

```python
from wordcloud import WordCloud

font_path = 'c:/Windows/Fonts/malgun.ttf'
```

```python
wordcloud = WordCloud(
    font_path = font_path, 
    background_color = 'white', # 배경색
    max_words = 10, # 사용할 워드의 개수
    width = 600,
    height = 600
    ).generate_from_frequencies(tags_total_selected)

# 그래프 크기 지정
plt.figure(figsize = (20, 15))

# 그래프 보여주기
plt.imshow(wordcloud)

# x축, y축 보이기
plt.axis('off') 

# 이미지로 저장하기
plt.savefig('./jeju_tag_wordcloud.png')
```
