```python
report = report.replace('\n', '')
report
```

#### - 명사 추출하기

```python
word_list = word.nouns(report)
word_list
```

#### - 단어 빈도 수 체크하기

```python
word_count = Counter(word_list)
word_count
```

 #### - 1자리 단어 제거하기

```python
word_list_finally = []

for i in word_list:
    if len(i) > 1:
        word_list_finally.append(i)
        
print(word_list_finally)    
```

```python
a = [i for i in word_list if len(i) > 1]
a
```

```python
word_count_finally = Counter(word_list_finally)
word_count_finally
```

```python
wordcloud = WordCloud(
    font_path = font_path, 
    background_color = 'white', # 배경색
    max_words = 30, # 사용할 워드의 개수
    width = 300,
    height = 150
    ).generate_from_frequencies(word_count_finally)

# 그래프 크기 지정
plt.figure(figsize = (15, 10))

# 그래프 보여주기
plt.imshow(wordcloud)

# x축, y축 보이기
plt.axis('off') 

# 이미지로 저장하기
plt.savefig('./report_wordcloud.png')
```

# 2. 강아지 모양 그리기

```python
import stylecloud
```

```python
stylecloud.gen_stylecloud(
    text = word_count_finally,
    
    # 워드 클라우드 스타일 지정
    icon_name = 'fas fa-dog',
    
    # 사용할 색상
    palette = 'colorbrewer.diverging.Spectral_11',
    
    # 배경색
    background_color = 'black',
    
    # 수직구조로 색상전환
    gradient = 'horizontal',
    
    # 폰트 지정
    font_path = font_path,
    
    # 이미지로 저장
    output_name = './fa_dog_wordcloud.png'
)
```
