# 웹 크롤링 - Melon

## 1. 웹 크롤링

## BeautifulSoup  방식

```python
driver = webdriver.Chrome('C:/ChromeDriver_exe/Chrome_99_driver.exe')
url = 'http://www.melon.com/chart/index.htm'
driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# 순위 top 10
songs = soup.select('tr')[1 : 11]

i = 0
for song in songs:
    title = song.select('div.wrap > div.wrap_song_info > div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.wrap > div.wrap_song_info > div.ellipsis.rank02 > a')[0].text
    i += 1
    print('순위[{}], 제목[{}], 가수[{}]'.format(i, title, singer))
```

## selenium  방식

```python
driver = webdriver.Chrome('C:/ChromeDriver_exe/Chrome_99_driver.exe')
url = 'http://www.melon.com/chart/index.htm'
driver.get(url)

# 순위 top 10
songs = driver.find_elements_by_css_selector('tr')[1 : 11]

i = 0
for song in songs:
    title = song.find_elements_by_css_selector('div.wrap > div.wrap_song_info > div.ellipsis.rank01 > span > a')[0].text
    singer = song.find_elements_by_css_selector('div.wrap > div.wrap_song_info > div.ellipsis.rank02 > a')[0].text
    i += 1
    print('순위[{}], 제목[{}], 가수[{}]'.format(i, title, singer))
```
