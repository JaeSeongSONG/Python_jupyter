```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('C:/ChromeDriver_exe/Chrome_99_driver.exe')
```

```python
url = 'https://www.yna.co.kr/news'
driver.get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
html = driver.page_source

html
```

```python
soup = BeautifulSoup(html, 'html.parser')

soup
```

```python
news = soup.select('div.slick-track')

news
```

```python
new = news[0]

new
```

```python
fin_news = new.select('strong')

fin_news[0].text
```

```python
for news in fin_news:
    print(news.text)
```
