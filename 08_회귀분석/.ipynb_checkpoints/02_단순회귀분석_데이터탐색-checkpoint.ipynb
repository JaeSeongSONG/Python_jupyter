### 1.3 이상치 제거하기

```python
df.isnull().sum()
```

```python
# horse power 가 나오지 않는 것을 확인

df.describe() 
```

```python
# unique()

df['horsepower'].unique() # '?' 값이 있는 것 확인
```

```python
df[df['horsepower'] == '?']
```

```python
# 이상치 제거
# 결측치로 바꾸고 가져오기

import numpy as np

df['horsepower'].replace('?', np.nan, inplace = True)
df['horsepower'].unique()
df.dropna(subset = ['horsepower'], axis = 0, inplace = True)
df
```

```python
# 이상치 제거
# 특정 컬럼만 조회하기

df = df[df['horsepower'] != '?']
df['horsepower'].unique()
```

```python
df.info()
```

### 1.4 horsepower 열 타입 바꾸기

```python
df['horsepower'] = df['horsepower'].astype('float')
df.info()
```

```python
df.to_csv('./file_new/auto-mpg_v2.csv', index = False)
```
