```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./file_new/auto-mpg_v2.csv')
df
```

# 1. 지도학습 모델
## 1. 단순회귀분석(선형회귀모델, 선형회귀예측모델)
#### - 종속변수가 있음
#### - 직선의 방정식을 사용 (y = ax + b)
#### - y = 종속변수,  x = 독립변수, a = 기울기, b = y절편
#### - 학습 : fit 함수

```python
# 사용할 변수들: mpg, cylinder, horsepower, weight

ndf = pd.DataFrame({'mpg' : df['mpg'],
                    'cylinders' : df['cylinders'],
                    'horsepower' : df['horsepower'],
                    'weight' : df['weight']})

ndf
```

### 1.1 산점도 그래프 그리기

```python
plt.scatter(x = ndf['weight'], y = ndf['mpg'], color = 'r')
plt.grid(True)
plt.show()
```

```python
ndf.plot(kind = 'scatter', x = 'weight', y = 'mpg', s = 10, figsize = (10, 5), color = 'r')
plt.grid(True)
plt.show()
```

#### - 중량이 커질수록 연비는 낮아지는 음의 상관관계를 갖는다.

```python
plt.subplot(1, 2, 1) 
plt.scatter(x = ndf['weight'], y = ndf['mpg'], color = 'red')
plt.grid(True)

plt.subplot(1, 2, 2)
sns.regplot(x = 'weight', y = 'mpg', color = 'red', data = ndf)
plt.grid(True)

plt.show()
```

### 1.2 회귀선 그래프 그리기

```python
fig = plt.figure(figsize = (15, 10))

# 첫번째 그래프 공간 만들기

ax1 = fig.add_subplot(1, 2, 1)

# 두번째 그래프 공간 만들기

ax2 = fig.add_subplot(1, 2, 2)

# 그래프 넣기

sns.regplot(x = 'weight',
            y = 'mpg',
            ax = ax2,
            color = 'red',
            fit_reg = True,
            data = ndf)

sns.regplot(x = 'weight',
            y = 'mpg',
            ax = ax1,
            color = 'Blue',
            fit_reg = False,
            data = ndf)

plt.show()
```

```python
sns.jointplot(x = 'weight',
              y = 'mpg',
              color = 'Green',
              data = ndf)

sns.jointplot(x = 'weight',
              y = 'mpg',
              kind = 'reg',
              color = 'Red',
              data = ndf)
```

```python
sns.jointplot(x = 'cylinders',
              y = 'horsepower',
              color = 'Green',
              data = ndf)

sns.jointplot(x = 'cylinders',
              y = 'horsepower',
              kind = 'reg',
              color = 'Red',
              data = ndf)
```

```python
sns.pairplot(ndf,
             markers = '+',
             kind = 'reg',)
plt.show()
```
