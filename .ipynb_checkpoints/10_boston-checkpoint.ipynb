# 보스턴 주택 가격 예측

## 1. 보스턴 주택 가격 데이터셋

```python
import numpy as np
import pandas as pd
import sklearn
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston_df = pd.DataFrame( boston.data, columns = boston.feature_names )
boston_df.head()
```

```python
# crim : 지역별 범죄 발생률
# zn   : 25000 평방피트를 초과하는 거주 지역 비율
# indus: 비상업 지역의 넓이 비율
# chas : 찰스강의 더미변수( 1은 경계, 0은 경계 아님 )
# nox  : 일산화질소 농도
# rm   : 거주할 수 있는 방 개수
# age  : 1940년 이전에 건축된 주택 비율
# dis  : 5개 주요 고용센터까지 가중 거리
# rad  : 고속도로 접근 용이도
# tax  : 10000달러당 재산세 비율
# ptratio: 지역의 교사와 학생 수 비율
# b    : 지역의 흑인 거주 비율
# lstat: 하위 계층의 비율
# price(medv): 본인 소유 주택 가격의 중앙값
```

## 2. 종속변수인 PRICE 컬럼 추가

```python
boston_df['PRICE'] = boston.target
boston_df.head()
```

#### - 데이터셋 크기 확인

```python
boston_df.shape
```

```python
boston_df.info()
```

## 3. 선형 회귀를 이용해 분석 모델 구축하기

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

#### - X, Y 분할하기

```python
Y = boston_df[ 'PRICE' ]
X = boston_df.drop( [ 'PRICE' ], axis = 1, inplace = False )
```

#### - 훈련용 데이터와 평가용 데이터 분할

```python
# train 데이터와 test 데이터를 7대 3의 비율로 분할

X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 0.3, random_state = 1 )
```

#### - 선형 회귀 분석 모델 생성

```python
lr = LinearRegression()
```

```python
lr.fit( X_train, Y_train )
```

#### - 평가 데이터에 대한 예측 수행 -> 결과값 구하기

#### - 회귀분석은 지도학습이므로 X_test에 대한 Y_test값을 이미 알고 있음

```python
Y_predict = lr.predict( X_test )
```

```python
Y_predict
```

## 4. 평가
#### - MSE: Y_test와 Y_predict의 오차를 계산하여 모델 평가

```python
mse = mean_squared_error( Y_test, Y_predict )
rmse = np.sqrt( mse )

print( 'MSE: {0:,.3f}, RMSE: {1:.3f}'.format( mse, rmse ) )
print( 'R^2( Variance score ) : {0:,.3f}'.format( r2_score( Y_test, Y_predict ) ) )
```

#### - 선형 회귀의 Y절편과 각 피처의 회귀계수를 확인
#### - 회귀 계수 값을 소수점 아래 2의자리에서 반올림

```python
print( 'Y 절편값: ', lr.intercept_ )
print( '회귀 계수 값: ', np.round( lr.coef_, 1 ) )
```

#### - 회귀 계수 값과 X의 feature_name을 묶어서 Series로 만들기
#### - 내림차순 정렬

```python
coef = pd.Series( data = np.round( lr.coef_, 2 ), index = X.columns )
coef.sort_values( ascending = False )
```

## - 5. 회귀 분석 결과를 산점도와 선형 회귀 그래프로 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 5행 3열의 subplot 생성

fig, axs = plt.subplots( figsize = ( 16, 16 ), ncols = 3, nrows = 5 )

# X_features 값에 X의 feature값들을 list화

x_features = [ 'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT' ]

# x_features의 각 원소에 인덱스 값을 붙여 regplot에 할당

for i, feature in enumerate( x_features ):
    row = int( i / 3 )
    col = i % 3
    sns.regplot( x = feature, y = 'PRICE', data = boston_df, ax = axs[row][col] )
```
