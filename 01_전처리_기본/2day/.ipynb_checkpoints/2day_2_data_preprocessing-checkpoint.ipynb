### 1.8 데이터 합치기

#### 행 단위로 합치기

```python
s1 = pd.Series([0, 1], index = ['A', 'B'])
s2 = pd.Series([2, 3, 4], index = ['a', 'b', 'c'])      
```

```python
print(s1)
```

```python
print(s2)
```

```python
pd.concat([s1, s2], axis = 0)
```

#### 컬럼 단위로 합치기

#### - CONCAT

```python
df_temp1 = pd.DataFrame([['Dog', 3], ['Cat', 4], ['Tiger', 8], ['Monkey', 5]],
                       index = [0, 1, 2, 3],
                       columns = ['animal', 'age'])
```

```python
df_temp1
```

```python
df_temp2 = pd.DataFrame([['집', 0], ['초원', 1], ['수풀', 2], ['초원', 1]],
                       index = [0, 1, 2, 3],
                       columns = ['area', 'times'])
```

```python
df_temp2
```

```python
pd.concat([df_temp1, df_temp2], axis = 1)
```

#### - MERGE

```python
df_info1 = pd.DataFrame({'고객번호' : [1001, 1002, 1003, 1004, 1005, 1006, 1007], 
                         '이름' : ['강', '김', '홍', '이', '박', '최', '송']},
                         columns = ['고객번호', '이름'],
                         index = [0, 1, 2, 3, 4, 5, 6])
```

```python
df_info1
```

```python
df_info2 = pd.DataFrame({'고객번호' : [1001, 1002, 1001, 1004, 1007, 1006, 1007],
                         '금액' : [10000, 20000, 30000, 40000, 50000, 60000, 70000]},
                         columns = ['고객번호', '금액'])
```

```python
df_info2
```

```python
pd.merge(df_info1, df_info2, how = 'inner')
```

```python
pd.merge(df_info1, df_info2, how = 'outer')
```

```python
pd.merge(df_info1, df_info2, how = 'left')
```

```python
pd.merge(df_info1, df_info2, how = 'right')
```
