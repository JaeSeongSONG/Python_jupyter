# 데이터 처리

## 1. 데이터 처리

### 1.1 결측값

#### - 데이터 누락값
#### - 결측값이 존재하면 변수 간의 관계가 왜곡될 수 있어서 분석의 정확성이 떨어짐
#### - pandas에서는 결측값을 NaN(Not a Number)으로 표기(None도 포함)
#### - 결측치는 제거나 대체로 처리함

# ---------------------------------------------------------------------------------------------
#### - isnull(), notnull()
#### - 컬럼별 결측값 개수 구하기: df.isnull().sum()
#### - 행별 결측값 개수 구하기: df.isnull().sum(1)
#### - 행 삭제: dropna(axis = 0)
#### - 열 삭제: dropna(axis = 1)
#### - 특정 값으로 채우기: df.fillna(0)
#### - 문자열로 채우기: df.fillna(' ')
#### - 변수별 평균으로 대체: df.fillna(df.mean())
