## 9. 반복문을 이용하여 파일 불러오기

```python
file_path = list()

for i in range(2010, 2020):
    for j in range(1, 13):
        file_path.append(f'./files/kto_{i}{str(j).zfill(2)}.xlsx')
        
for file_name in file_path:
    df = pd.read_excel(file_name, header = 1, skipfooter = 4, usecols = 'A : G')
    print(len(df))   
```

```python
df_all = pd.DataFrame()

for yy in range(2010, 2020):
    for mm in range(1, 13):
        
        try:
            
            file_path = './files/kto_{}{}.xlsx'.format(yy, str(mm).zfill(2))
            df = pd.read_excel(file_path, header = 1, skipfooter = 4, usecols = 'A:G')

            # 기준년월 column 추가
            df['기준년월'] = str(yy) + '-' + str(mm).zfill(2)

            # 특정 행 제거
            continentis_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계']
            condition = df['국적'].isin(continentis_list)
            df_country = df[condition == False]

            # index 재설정
            df_country_newindex = df_country.reset_index(drop = True)

            # 대륙 column 추가 후 구분하기
            df_country_newindex['대륙'] = ['아시아'] * 25 + ['아메리카'] * 5 + ['유럽'] * 23 + ['오세아니아'] * 3 + \
                                          ['아프리카'] * 2 + ['기타대륙'] + ['교포']

            # 관광객 비율 column 추가
            df_country_newindex['관광객비율(%)'] = round(df_country_newindex['관광'] / df_country_newindex['계'] * 100, 1)

            # 전체 비율 column 추가
            df_country_newindex['전체비율(%)'] = round(df_country_newindex['관광'] / \
                                                 sum(df_country_newindex['관광']) * 100, 1)
            
            # 통합하기
            df_all = df_all.append(df_country_newindex, ignore_index = True)
            
        except:
            
            pass
        
print(df_all)      
```

```python
def create_kto_data(yy, mm):
    
    file_path = './files/kto_{}{}.xlsx'.format(yy, str(mm).zfill(2))
    df = pd.read_excel(file_path, header = 1, skipfooter = 4, usecols = 'A:G')

    # 기준년월 column 추가
    df['기준년월'] = str(yy) + '-' + str(mm).zfill(2)

    # 특정 행 제거
    continentis_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계']
    condition = df['국적'].isin(continentis_list)
    df_country = df[condition == False]

    # index 재설정
    df_country_newindex = df_country.reset_index(drop = True)

    # 대륙 column 추가 후 구분하기
    df_country_newindex['대륙'] = ['아시아'] * 25 + ['아메리카'] * 5 + ['유럽'] * 23 + ['오세아니아'] * 3 + \
                                  ['아프리카'] * 2 + ['기타대륙'] + ['교포']

    # 관광객 비율 column 추가
    df_country_newindex['관광객비율(%)'] = round(df_country_newindex['관광'] / df_country_newindex['계'] * 100, 1)

    # 전체 비율 column 추가
    df_country_newindex['전체비율(%)'] = round(df_country_newindex['관광'] / \
                                          sum(df_country_newindex['관광']) * 100, 1)
    return df_country_newindex
```

```python
for yy in range(2010, 2020):
    for mm in range(1, 13):
        try:
            
            temp = create_kto_data(yy, mm)
            df_all = df_all.append(temp, ignore_index = True)
            
        except:
            
            pass
        
print(df_all)
```
