```python
from matplotlib import font_manager,rc
import platform
```

#### - 운영체제별 폰트 정의하기

```python
if platform.system() == 'Windows' :
    path = 'c:/Windows/Fonts/Malgun.ttf'
    font_name = font_manager.FontProperties(fname = path).get_name()
    rc('font', family = font_name)

elif platform.system() == 'Darwin': 
    rc('font', family = 'AppleGothic')

else :
    print('Check Your OS System...')
```

## 1. 데이터 불러오기

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./files/kto_total.xlsx')
df
```

## 2. 데이터 탐색

```python
df_filter = df[df['국적'] == '중국']
df_filter
```

## 3. 그래프 그리기

```python
plt.plot(df_filter['기준년월'], df_filter['관광'])
plt.show()
```

#### - xticks: 축 눈금 설정 함수

```python
plt.figure(figsize = (12, 4))
plt.plot(df_filter['기준년월'], df_filter['관광'])
plt.title("중국 국적에 대한 관광객 추이 분석")
plt.xlabel('기준년월')
plt.ylabel('관광객수')

plt.xticks(['2010-01', '2011-01', '2012-01', '2013-01', '2014-01', '2015-01', '2016-01', '2017-01', '2018-01', '2019-01'])

plt.show()
```

```python
df.sort_values(['전체비율(%)'], ascending = False).head()
```

```python
cntry_list = ['중국', '일본', '대만', '미국', '홍콩']
```

```python
for cntry in cntry_list :
    
    condition = (df["국적"] == cntry)

    df_filter = df[condition]

    # 그래프 꾸미기
    # 그래프 크기 조절

    plt.figure(figsize = (12, 4))

    # 그래프에 데이터 넣기..
    plt.plot(df_filter["기준년월"], df_filter["관광"])

    # 그래프 제목 넣기
    plt.title("{}에 대한 관광객 추이 분석".format(cntry))
    # x축에 대한 라벨 추가
    plt.xlabel("기준년월")
    # y축에 대한 라벨 추가
    plt.ylabel("관광객수")

    # x축에 대한 눈금을 조정..
    plt.xticks(["2010-01", "2011-01", "2012-01", "2013-01", "2014-01",
                "2015-01", "2016-01", "2017-01", "2018-01", "2019-01"])

    plt.show()
```

```python
df_japan = df[df['국적'] == '일본']
df_taiwan = df[df['국적'] == '대만']
df_us = df[df['국적'] == '미국']
df_hongkong = df[df['국적'] == '홍콩']

plt.figure(figsize = (17, 7))
plt.plot(df_japan['기준년월'], df_japan['관광'], label = '일본')
plt.plot(df_taiwan['기준년월'], df_taiwan['관광'], label = '대만')
plt.plot(df_us['기준년월'], df_us['관광'], label = '미국')
plt.plot(df_hongkong['기준년월'], df_hongkong['관광'], label = '홍콩')

plt.title('국적별 관광객 추이 분석')
plt.xlabel('기준년월')
plt.ylabel('관광객수')
plt.legend()

plt.xticks(['2010-01', '2011-01', '2012-01', '2013-01', '2014-01', '2015-01', '2016-01', '2017-01', '2018-01', '2019-01'])
plt.show()
```

## 4. 히트맵 그리기

```python
df['연도'] = df['기준년월'].str.slice(0, 4)
df['월'] = df['기준년월'].str.slice(5, 7)
df
```

```python
df_filter = df[(df['국적'] == '중국')]
df_filter
```

### 4.1 Pivot 데이터 만들기

```python
df_pivot = df_filter.pivot_table(values = '관광',
                     index = '연도',
                     columns = '월')
df_pivot
```

### 4.2 히트맵 그리기

```python
import seaborn as sns
```

#### - 데이터 넣기

```python
plt.figure(figsize = (16, 10))
sns.heatmap(df_pivot, annot = True, fmt = '.0f', cmap = 'rocket_r')
plt.title('년/월별 [중국] 관광객 히트맵 시각화분석')
plt.show()
```

```python
for cntry in cntry_list :
    
    condition = (df["국적"] == cntry)

    df_filter = df[condition]
    
    df_pivot = df_filter.pivot_table(values = '관광',
                     index = '연도',
                     columns = '월')

    plt.figure(figsize = (16, 10))
    sns.heatmap(df_pivot, annot = True, fmt = '.0f', cmap = 'rocket_r')
    plt.title(f'년/월별 {cntry} 관광객 히트맵 시각화분석')
    plt.show()
```
