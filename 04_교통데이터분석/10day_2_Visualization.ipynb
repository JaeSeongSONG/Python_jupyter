### 1.1 기준월 및 기준일자 별 버스 이용량 분석 비교

#### - 사용할 분석요소: 기준월, 기준일, 승객구분
#### - 사용할 함수: count()

```python
df_bus_card_tot[df_bus_card_tot['승객구분'] == 'None'] 
```

```python
df_pivot = df_bus_card_tot.pivot_table(values = '승객구분',
                             index = '기준월',
                             columns = '기준일',
                             aggfunc = 'count')

df_pivot
```

#### - 히트맵 그리기(Nan값은 0으로)

```python
df_pivot.fillna(0, inplace = True)

plt.figure(figsize = (16, 10))
sns.heatmap(df_pivot, annot = True, fmt = '.0f', cmap = 'rocket_r')
plt.title('기준월 및 기준일자 별 버스 이용량 분석 비교')
plt.show()
```

### 1.2 기준일 및 시간대 별 버스 이용량 분석 비교

```python
df_pivot = df_bus_card_tot.pivot_table(values = '승객구분',
                             index = '기준일',
                             columns = '기준시간',
                             aggfunc = 'count')

df_pivot
```

```python
df_pivot.fillna(0, inplace = True)

plt.figure(figsize = (16, 10))
sns.heatmap(df_pivot, annot = True, fmt = '.0f', cmap = 'rocket_r')
plt.title('기준일 및 시간대 별 버스 이용량 분석 비교')
plt.show()
```

### 1.3 기준시간 및 분 별 버스 이용량 분석 비교

```python
df_pivot = df_bus_card_tot.pivot_table(values = '승객구분',
                             index = '기준시간',
                             columns = '기준시간(분)',
                             aggfunc = 'count')

df_pivot
```

```python
df_pivot.fillna(0, inplace = True)

plt.figure(figsize = (16, 10))
sns.heatmap(df_pivot, annot = True, fmt = '.0f', cmap = 'rocket_r')
plt.title('기준시간 및 분 별 버스 이용량 분석 비교')
plt.show()
```

### 1.4 기준일 및 시간대 별 버스 이용량 분석 비교

```python
df_pivot = df_bus_card_tot.pivot_table(values = '버스내체류시간(분)',
                             index = '기준일',
                             columns = '기준시간',
                             aggfunc = 'mean')

df_pivot
```

```python
df_pivot.fillna(0, inplace = True)

plt.figure(figsize = (20, 15))
sns.heatmap(df_pivot, annot = True, fmt = '.0f', cmap = 'rocket_r')
plt.title('기준일 및 시간대 별 버스내 체류시간(분) 분석 비교')
plt.show()
```

### 1.5 시간대별 승객구분별 버스 내 체류시간(분) 분석

```python
df_temp = pd.DataFrame()

df_temp['기준시간'] = df_bus_card_tot['기준시간']
df_temp['승객구분'] = df_bus_card_tot['승객구분']
df_temp['버스내체류시간(분)'] = df_bus_card_tot['버스내체류시간(분)']

df_temp
```

#### - None 값이 있다는 것을 확인

```python
df_temp.drop_duplicates(['승객구분'])
```

```python
df_temp['승객구분'].unique()
```

#### - transpose() 옆으로 눕히기

```python
df_temp2 = (df_temp.groupby(['기준시간', '승객구분'], as_index = False).sum()). \
                                                         sort_values(by = ['버스내체류시간(분)'], ascending = False)

df_temp2.head(10).transpose()
```

```python
fig = plt.figure(figsize = (25, 10))
sns.barplot(x = '기준시간', y = '버스내체류시간(분)', hue = '승객구분', data = df_temp2)
plt.title('시간대별 승객구분별 버스 내 체류시간(분) 분석')
plt.show()
```

```python
fig = plt.figure(figsize = (25, 10))
sns.barplot(x = '기준시간', y = '버스내체류시간(분)', hue = '승객구분', data = df_temp)
plt.title('시간대별 승객구분별 버스 내 체류시간(분) 분석')
plt.show()
```

### 1.5 시간 및 승객구분별 빈도 분석

```python
plt.figure(figsize = (12, 4))
g = sns.histplot(data = df_temp2, 
                 x = '기준시간', 
                 bins = 30, 
                 kde = True, 
                 hue = '승객구분', 
                 multiple = 'stack', 
                 stat = 'density', 
                 shrink = 0.6)

plt.title('시간 및 승객구분별 빈도분석')
plt.xlabel('기준시간')
plt.ylabel('rate')
plt.show()
```

### 1.6 승하차정류장 및 버스내체류시간(분)별 상위 30건 분석

```python
df_temp = pd.DataFrame()

df_temp['승하차정류장'] = df_bus_card_tot['승차정류장'] + ' --> ' + df_bus_card_tot['하차정류장'] 
df_temp['승객구분'] = df_bus_card_tot['승객구분']
df_temp['버스내체류시간'] = df_bus_card_tot['버스내체류시간(분)']

df_temp
```

```python
df_temp_sel = (df_temp.groupby(['승하차정류장', '승객구분'], as_index = False).sum()). \
                                                         sort_values(by = ['버스내체류시간'], ascending = False).head(30)

df_temp_sel
```

```python
plt.figure(figsize = (12, 4))
plt.plot(df_temp_sel['승하차정류장'], df_temp_sel['버스내체류시간'])
plt.title('승하차정류장별 버스내 체류시간(분) 상위 30건 분석')
plt.xlabel('승하차정류장')
plt.ylabel('버스내체류시간')
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```
