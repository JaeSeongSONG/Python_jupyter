## 1. 컬럼명 한글명으로 변환

```python
trfcard = pd.read_csv('./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard (0)/trfcard.csv')
trfcard
```

```python
trfcard_col = pd.read_excel('./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard (0)/trfcard_columns.xlsx', \
                               sheet_name = '컬럼 정의서', \
                               usecols = 'B, C',
                               header = 2)

trfcard_col
```

```python
trfcard_col.to_dict('records')
```

```python
trfcard_col_dict = {}

for col in trfcard_col.to_dict('records'):
    trfcard_col_dict[col['컬럼명 (영문)']] = col['컬럼명 (한글)']
    
trfcard_col_dict
```

```python
trfcard_col_dict_kor = trfcard.rename(columns = trfcard_col_dict)
trfcard_col_dict_kor
```
