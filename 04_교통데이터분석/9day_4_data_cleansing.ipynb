```python
import pandas as pd
```

```python
df_bus_kor = pd.DataFrame()

for i in range(100):
    # 교통 데이터 불러오는 거
    bus_file_path = './01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard ({})/trfcard.csv'.format(i)
    df_bus = pd.read_csv(bus_file_path)
    
    # 영문 / 한글 컬럼명 정의서 데이터 불러오기
    col_name_file_path = './01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard ({})/trfcard_columns.xlsx'.format(i)
    df_col_name = pd.read_excel(col_name_file_path,
                                usecols = 'B, C',
                                header = 2)
    
    df_col_name = df_col_name.to_dict('records')
    temp_dict = {}
    for name in df_col_name:
        temp_dict[name['컬럼명 (영문)']] = name['컬럼명 (한글)']

    df_bus_kor_temp = df_bus.rename(columns = temp_dict)
    
    df_bus_kor = df_bus_kor.append(df_bus_kor_temp, ignore_index = True)
```
