```python
from scipy import stats
from statsmodels.formula.api import ols, glm
```

```python
red_wine_quality = wine.loc[ wine.type == 'red', 'quality' ]
red_wine_quality
```

```python
# '종속변수~ 독립변수(연속형데이터)'

Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + \
            free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
```

```python
# ols: 최소자승법, 첫번째 인자로 변수, 두번째 인자로 실제 사용할 변수값을 가진 데이터프레임
#      근사적으로 구하려는 해와 실제 해의 오차의 제곱의 합이 최소가 되는 해를 구하는 방법
#      종속변수의 분산을 독립변수로 설명하는 과정

# Df Residuals: DF는 Degree of Freedom으로 자유도를 뜻하는데, DF Residuals는 전체 표본 수에서 측정되는 변수들
#              (종속변수 및 독립변수)의 개수를 빼서 구한다. 

# Df Model: 독립변수의 개수이다.

# R-squared: 설명된 분산 / 종속변수의 전체 분산
#            1에 가까울수록 좋음

# p-value = P > |t|: 독립변수들의 유의확률, 0.05보다 작아야 유리
#                    어떤 사건이 우연히 발생할 확률


regression_result = ols( Rformula, data = wine ).fit()
```

```python
regression_result.summary()
```

```python
sample1 = wine[ wine.columns.difference( [ 'type', 'quality' ] ) ]
sample1 = sample1[ 0 : 5 ]
sample1
```

```python
sample1_pred = regression_result.predict( sample1 )
sample1_pred
```
