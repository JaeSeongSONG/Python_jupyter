```python
import pandas as pd
```

```python
file_path = './files_new/df_20TABLE_kor_명세서일반내역.xlsx'
df_20TABLE_org = pd.read_excel(file_path)
df_20TABLE_org
```

```python
print(df_20TABLE_org['수진자연령'])
print(df_20TABLE_org['요양개시일자'])
print(df_20TABLE_org['심결본인부담금'])
```

## 2. 데이터프레임 추출해서 새로 만들기

```python
df_month_age = pd.DataFrame({'수진자연령':df_20TABLE_org['수진자연령'],
                             '요양개시일자':df_20TABLE_org['요양개시일자'],
                             '심결본인부담금':df_20TABLE_org['심결본인부담금']})

df_month_age
```

## 3. 결측치 제거

```python
df_month_age.info()
```

#### - 결측치가 없는 것으로 판단

##  4. 조건 만들기

#### - 연령대 별 조건 만들기

```python
df_month_age[df_month_age['수진자연령'] <= 9]
df_month_age[(df_month_age['수진자연령'] < 20) & (df_month_age['수진자연령'] >= 10)]
df_month_age[(df_month_age['수진자연령'] < 30) & (df_month_age['수진자연령'] >= 20)]
df_month_age[(df_month_age['수진자연령'] < 40) & (df_month_age['수진자연령'] >= 30)]
df_month_age[(df_month_age['수진자연령'] < 50) & (df_month_age['수진자연령'] >= 40)]
df_month_age[(df_month_age['수진자연령'] < 60) & (df_month_age['수진자연령'] >= 50)]
df_month_age[(df_month_age['수진자연령'] < 70) & (df_month_age['수진자연령'] >= 60)]
df_month_age[(df_month_age['수진자연령'] < 80) & (df_month_age['수진자연령'] >= 70)]
df_month_age[(df_month_age['수진자연령'] < 90) & (df_month_age['수진자연령'] >= 80)]
df_month_age[(df_month_age['수진자연령'] < 100) & (df_month_age['수진자연령'] >= 90)]
df_month_age[df_month_age['수진자연령'] >= 100]
```

## 5. 연령대 별 데이터 만들기

```python
df_9under = df_month_age[df_month_age['수진자연령'] <= 9]
df_9under = df_9under.reset_index(drop = True)
df_9under['연령대'] = '9세이하'
df_9under
```

```python
df_10 = df_month_age[(df_month_age['수진자연령'] < 20) & (df_month_age['수진자연령'] >= 10)]
df_10 = df_10.reset_index(drop = True)
df_10['연령대'] = '10대'
df_10
```

```python
df_20 = df_month_age[(df_month_age['수진자연령'] < 30) & (df_month_age['수진자연령'] >= 20)]
df_20 = df_20.reset_index(drop = True)
df_20['연령대'] = '20대'
df_20
```

```python
df_30 = df_month_age[(df_month_age['수진자연령'] < 40) & (df_month_age['수진자연령'] >= 30)]
df_30 = df_30.reset_index(drop = True)
df_30['연령대'] = '30대'
df_30
```

```python
df_40 = df_month_age[(df_month_age['수진자연령'] < 50) & (df_month_age['수진자연령'] >= 40)]
df_40 = df_40.reset_index(drop = True)
df_40['연령대'] = '40대'
df_40
```

```python
df_50 = df_month_age[(df_month_age['수진자연령'] < 60) & (df_month_age['수진자연령'] >= 50)]
df_50 = df_50.reset_index(drop = True)
df_50['연령대'] = '50대'
df_50
```

```python
df_60 = df_month_age[(df_month_age['수진자연령'] < 70) & (df_month_age['수진자연령'] >= 60)]
df_60 = df_60.reset_index(drop = True)
df_60['연령대'] = '60대'
df_60
```

```python
df_70 = df_month_age[(df_month_age['수진자연령'] < 80) & (df_month_age['수진자연령'] >= 70)]
df_70 = df_70.reset_index(drop = True)
df_70['연령대'] = '70대'
df_70
```

```python
df_80 = df_month_age[(df_month_age['수진자연령'] < 90) & (df_month_age['수진자연령'] >= 80)]
df_80 = df_80.reset_index(drop = True)
df_80['연령대'] = '80대'
df_80
```

```python
df_90 = df_month_age[(df_month_age['수진자연령'] < 100) & (df_month_age['수진자연령'] >= 90)]
df_90 = df_90.reset_index(drop = True)
df_90['연령대'] = '90대'
df_90
```

```python
df_month_age_any = pd.concat([df_9under, df_10, df_20, df_30, df_40, df_50, df_60, df_70, df_80, df_90])
df_month_age_any = df_month_age_any.reset_index(drop = True)
df_month_age_any
```

## 6. 요양개시일자 처리하기

```python
df_month_age_any['요양개시일자']
```

```python
df_month_age_any['요양개시일자'] = df_month_age_any['요양개시일자'].astype(str)
```

```python
df_month_age_any['요양개시월'] = df_month_age_any['요양개시일자'].str[4 : 6]
df_month_age_any['요양개시월']
```

## 7. 저장하기

```python
df_month_age_any.to_excel('./files_new/df_month_age_any_월별연령대별본인부담비용.xlsx', index = False)
```
