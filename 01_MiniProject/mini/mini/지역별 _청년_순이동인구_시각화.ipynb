```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns

plt.rc('font', family = 'Malgun Gothic')
```

```python
# 데이터 불러오기
path = 'C:/Users/admin/Desktop/미니 프로젝트 1/시각화용(전처리 완료)/10년간_지역별_청년인구_순이동.csv'
df = pd.read_csv(path)

# 컬럼명 변경
df = df.rename(columns = {'행정구역(시군구)별' : '지역'})
df
```

```python
df.info()
```

```python
# 히트맵 생성을 위해 피벗테이블 만들기
pt = df.pivot_table(values = ['2012 년', '2013 년', '2014 년','2015 년','2016 년', '2017 년', '2018 년', '2019 년', '2020 년', '2021 년'],
                    index = '지역')
pt
```

```python
# 히트맵 만들기
plt.figure(figsize = (16, 10))
sns.heatmap(pt, annot = True, fmt = '.0f', cmap = 'rocket_r')
plt.title('지역별 청년층 순이동 인구 수')
plt.show()
```
