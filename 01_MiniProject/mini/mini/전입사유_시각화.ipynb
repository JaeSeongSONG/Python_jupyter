```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
```

```python
# 데이터 불러오기
path = 'C:/Users/admin/Desktop/미니 프로젝트 1/시각화용(전처리 완료)/20년_21년_청년층_경기도_전입_사유_및_전입_합계(비율).csv'
df = pd.read_csv(path).drop(['Unnamed: 0'], axis = 1)
```

```python
# 계 행만 안 씀
df = df[(df['전입_사유'] != '계')]
df
```

```python
# 그래프 크기
plt.figure(figsize=(10,10))

# 중심에서 벗어나는 정도 설정하기
explode = list()
for i in range(len(df)) :
    explode.append(0.1)

plt.pie(df['2020년_청년층_전입_합계'], labels = df['전입_사유'], autopct='%.1f%%', explode = explode)
plt.title('20년 청년층 전입 사유')

plt.show()
```

```python
# 그래프 크기
plt.figure(figsize=(10,10))

# 중심에서 벗어나는 정도 설정하기
explode = list()
for i in range(len(df)) :
    explode.append(0.1)

plt.pie(df['2021년_청년층_전입_합계'], labels = df['전입_사유'], autopct='%.1f%%', explode = explode)
plt.title('21년 청년층 전입 사유')

plt.show()
```

```python
# 그래프 크기
plt.figure(figsize=(10,10))

# 중심에서 벗어나는 정도 설정하기
explode = list()
for i in range(len(df)) :
    explode.append(0.1)

plt.pie(df['청년층_전입_총_합계'], labels = df['전입_사유'], autopct='%.1f%%', explode = explode)
plt.title('20년, 21년 청년층 총 전입사유')

plt.show()
```
