```python
import pandas as pd
```

```python
# 데이터 불러오기
path = 'C:/Users/admin/Desktop/미니 프로젝트 1/전입전출인구.csv'
df = pd.read_csv(path)
df
```

```python
# 이상한 컬럼 제거
df = df.drop(['Unnamed: 14', '단위'], axis = 1)
df
```

```python
# 전국이라는 값을 가진 행은 필요없음
df = df[(df['행정구역(시군구)별'] != '전국')]
df
```

```python
# 청년층 인구만 씀
df = df[(df['연령별'] == '20 - 24세') | (df['연령별'] == '25 - 29세') | (df['연령별'] == '30 - 34세')]
df
```

```python
# 순이동(전입- 전출)인 행만 사용
df = df[(df['항목'] == '순이동[명]')]
df
```

```python
# 지역별로 묶은 후 더함
df = df.groupby('행정구역(시군구)별').sum()
df
```

```python
# 저장
df.to_csv('C:/Users/admin/Desktop/미니 프로젝트 1/전국_순이동인구.csv', index = True)
```
