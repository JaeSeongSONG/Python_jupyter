```python
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_csv('C:/Users/admin/Desktop/미니 프로젝트 1/데이터 통합용/전처리 필요/주거 관련/유형별_매매가격_대비_전세가격_비율.csv')
df
```

```python
# 주택유형이 '종합'인 행만 사용
df = df[(df['주택유형별(1)'] == '종합')]
df
```

```python
# 주택유형 컬럼은 제거
df = df.drop(['주택유형별(1)'], axis = 1)
df
```

```python
# 0번부터 6번 인덱스의 행은 제거
df_new = df.drop([0, 1, 2, 3, 4, 5, 6]).reset_index(drop = True)
df_new
```

```python
# 컬럼명 변경, 및 재배열
df_new = df_new.rename(columns = {'지역별(1)' : '지역', '2022.01' : '매매가격_대비_전세가격_비율'}).sort_values(by = '지역').\
         reset_index(drop = True)
df_new
```

```python
# 저장
df_new.to_csv('C:/Users/admin/Desktop/미니 프로젝트 1/데이터 통합용/매매가격_대비_전세가격_비율.csv')
```

```python
# 필요없는 컬럼 제거
df_new = df_new.drop(['지역'], axis = 1)
```

```python
# 합치기
df_all = pd.read_csv('C:/Users/admin/Desktop/미니 프로젝트 1/시각화용(전처리 완료)/지역별_데이터.csv')
```

```python
df_1 = pd.concat([df_all, df_new], axis = 1)
```

```python
df_1 = df_1.drop(['Unnamed: 0'], axis = 1)
```
