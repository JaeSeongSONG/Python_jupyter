```python
# 지역별 노년부양비 계산하기
# 변수명 df_loc(local)
path = 'C:/Users/admin/Desktop/미니 프로젝트 1/데이터 통합용/전처리 필요/인구 관련/1990년_2020년_주민등록인구.csv'
df_loc = pd.read_csv(path)
```

```python
# 사용하지 않은 컬럼 제거
df_loc = df_loc.drop(['항목', '성별', '단위'] , axis = 1)
```

```python
# 컬럼명 변경
df_loc = df_loc.rename(columns = {'행정구역(시군구)별' : '지역'})
```

```python
# 지역 컬럼 중 값이 전국인 행은 안 쓰고 인덱스 번호 재배열
# 사용하지 않는 컬럼(Unnamed: 34) 제거
df_loc = df_loc[(df_loc['지역'] != '전국')].reset_index(drop = True)
df_loc['연령별'].unique()
df_loc = df_loc.drop(['Unnamed: 34'], axis = 1)
```

```python
# 사용하지 않을 컬럼들 리스트에 넣기
list_not_age = ['계', '0 - 4세', '5 - 9세', '10 - 14세', '85 - 89세', '90 - 94세', '95 - 99세', '100세 이상']
```

```python
# 반복문 돌면서 사용하지 않을 컬럼리스트에 일치하면 빼기
for not_age in list_not_age :
    df_loc = df_loc[(df_loc['연령별'] != not_age)].reset_index(drop = True)
```

```python
# 결측치 0으로 바꿈
df_loc = df_loc.fillna(0)
```

```python
# 생산가능인구 데이터 만들기
# 노인 연령 리스트
list_old = ['65 - 69세', '70 - 74세', '75 - 79세', '80세 이상', '80 - 84세', '85세 이상']
df_loc_young = df_loc

# 생산가능인구(df_loc_young) 만들기
for old in list_old :
    df_loc_young = df_loc_young[(df_loc_young['연령별'] != old)].reset_index(drop = True)
```

```python
# 노인 데이터 만들기
df_loc_old = df_loc

# 노인(df_loc_old) 만들기 
df_loc_old = df_loc[(df_loc['연령별'] == '65 - 69세') | (df_loc['연령별'] == '70 - 74세') | (df_loc['연령별'] == '75 - 79세') |
                    (df_loc['연령별'] == '80세 이상')]
```

```python
# 인덱스 번호 재배열
df_loc_old = df_loc_old.reset_index(drop = True)
```

```python
# 지역별로 묶은 후 더하기
df_loc_young_sum = df_loc_young.groupby(['지역']).sum()
```

```python
# 지역별로 묶은 후 더하기
df_loc_old_sum = df_loc_old.groupby(['지역']).sum()
```

```python
# 확인
df_loc_young_sum
```

```python
# 확인
df_loc_old_sum
```

```python
# 임시 데이터 프레임 만들고 지역 컬럼 만들어서 값 넣음
df_res = pd.DataFrame()
df_res['지역'] = df_loc['지역'].unique()
```

```python
# 순서 재배열
df_res = df_res.sort_values(by = '지역', ascending = True)
```

```python
# 컬럼 리스트와 인덱스 리스트 만들기
col_list = df_loc_young_sum.columns
index_list = df_loc_young_sum.index

# 임시 데이터 프레임에 미리 값 넣음
for col in col_list :
    df_res[col] = 0
```

```python
# 임시 데이터 프레임의 인덱스 번호 재배열
df_res = df_res.set_index(keys=['지역'], inplace=False, drop=True)
```

```python
# 노인부양비 계산
# 반복문으로 모든 행, 열을 돌면서 일일이 값 넣음
for col in col_list :
    for index in index_list :
        ratio = round(df_loc_old_sum[col].loc[index] / df_loc_young_sum[col].loc[index] * 100, 2)
        print(col, index, ratio)
        df_res[col].loc[index] = ratio
```
