```python
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_csv('C:/Users/admin/Desktop/미니 프로젝트 1/데이터 통합용/전처리 필요/취업 관련/시도·산업·종사자규모별_사업체수__종사자수.csv',
                header = 1)
```

```python
df
```

```python
# 안 쓸 컬럼 제거
df = df.drop(['종사자수 (명)', '종사자규모별'], axis = 1)
df
```

```python
# 전국 행 제거
df = df[(df['행정구역별'] != '전국')]
df
```

```python
# 컬럼명 변경
df.columns = ['지역', '산업', '사업체수']
df
```

```python
# 전체 산업 행 제거
df = df[(df['산업'] != '전체 산업')]
df
```

```python
# 청년들이 선호하는 산업군 : 경영금융사무직, 교육법률관련직, 기술기능직
# 기피 산업군 리스트에 넣기
print(df['산업'].unique())
list_not_business = ['농업, 임업 및 어업(01~03)', '광업(05~08)', '제조업(10~34)', '전기, 가스, 증기 및 공기조절 공급업(35)',
                     '수도, 하수 및 폐기물 처리, 원료 재생업(36~39)', '건설업(41~42)', '운수 및 창고업(49~52)', 
                     '부동산업(68)', '사업시설 관리, 사업 지원 및 임대 서비스업(74~76)', '도매 및 소매업(45~47)',
                     '숙박 및 음식점업(55~56)', '협회 및 단체, 수리 및 기타 개인 서비스업(94~96)',
                     '보건업 및 사회복지 서비스업(86~87)', '교육 서비스업(85)']
```

```python
df_new = df
```

```python
# 기피 산업군 제외하기
for business in list_not_business :
    df_new = df_new[(df_new['산업'] != business)]
df_new
```

```python
# 지역별로 나누기
# 지역 이름 리스트
list_loc_name = df_new['지역'].unique()

# 지역별로 나눈 데이터 프레임을 저장할 딕셔너리
dict_df = dict()

# for문
for name in list_loc_name :
    dict_df[name] = df_new[(df_new['지역'] == name)]
```

```python
# 확인
dict_df
```

```python
# 각 지역에서 산업별로 묶기
for k, v in dict_df.items() :
    dict_df[k] = dict_df[k].groupby(['산업']).sum().sort_values(by = '사업체수', ascending = False)
```

```python
# 저장하기
for k, v in dict_df.items() :
    v.to_csv('C:/Users/admin/Desktop/미니 프로젝트 1/시각화용(전처리 완료)/지역별 사업체 수/{}_사업체_수.csv'.format(k))
```

```python
dict_df
```

```python
# 수도권과 지방의 선호 사업체수 비교해보기
# 사업체수를 지역별로 합치기
df_new = df_new. groupby(['지역']).sum()
df_new
```

```python
# 컬럼명 변경
df_new.columns = ['선호_사업군의_사업체_수']
```

```python
df_new
```

```python
# 수도권 : 서울, 경기, 인천
# 지방 : 나머지
locals = df['지역'].unique()

for local in locals :
    if local == '서울' or local == '경기' or local == '인천' :
        sudo += df_new.loc[local]
    else :
        jibang += df_new.loc[local]

df_tmp = pd.DataFrame()
df_tmp['지역'] = ['수도권', '지방']
df_tmp['선호_사업체_수'] = [sudo.sum(), jibang.sum()]
df_tmp
```
