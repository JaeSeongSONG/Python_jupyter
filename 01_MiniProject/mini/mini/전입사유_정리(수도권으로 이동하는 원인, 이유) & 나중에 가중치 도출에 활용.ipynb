```python
# 데이터 불러오기
path_20 = 'C:/Users/admin/Desktop/미니 프로젝트 1/인구 관련/2020년_경기_전입사유.xlsx'
path_21 = 'C:/Users/admin/Desktop/미니 프로젝트 1/인구 관련/2021년_경기_전입사유.xlsx'

df_20 = pd.read_excel(path_20, header = 5)
df_21 = pd.read_excel(path_21, header = 5)
```

```python
# 사용할 컬럼만 추출
df_20 = df_20[['Unnamed: 1', '20세 미만', '20대', '30대']]
df_21 = df_21[['Unnamed: 1', '20세 미만', '20대', '30대']]
```

```python
# 컬럼명 변경
df_20.columns = ['전입_사유', '20세 미만', '20대', '30대']
df_21.columns = ['전입_사유', '20세 미만', '20대', '30대']
```

```python
# 사용하지 않는 행 제거
# 인덱스 번호가 10번인 행 ~ 25번인 행만 씀
drop_row_list = list(range(10, 26))
# drop_row_list.append(0)
# drop_row_list.append(1)

df_20 = df_20.drop(drop_row_list).reset_index(drop = True)
df_21 = df_21.drop(drop_row_list).reset_index(drop = True)
```

```python
# 각 사유별 20세 미만, 20대, 30대 합치기
# 임시 데이터  프레임 만들고 미리 컬럼 생성해놓기
df_new = pd.DataFrame()
df_new['전입_사유'] = df_20['전입_사유']
df_new['2020년_청년층_전입_합계'] = 0
df_new['2021년_청년층_전입_합계'] = 0
df_new['청년층_전입_총_합계'] = 0

# 반복문으로 각 행 더해서 임시 데이터 프레임에 넣음
for i in range(8) :
    df_new['2020년_청년층_전입_합계'].iloc[i] = df_20.drop(['전입_사유'], axis = 1).iloc[i].sum()
    df_new['2021년_청년층_전입_합계'].iloc[i] = df_21.drop(['전입_사유'], axis = 1).iloc[i].sum()

# 20년, 21년 전입 인구를 더해서 새 컬럼에 넣음
for i in range(8) :
    df_new['청년층_전입_총_합계'].iloc[i] = df_new.drop(['전입_사유'], axis = 1).iloc[i].sum()
```
