```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
```

```python
# 데이터 불러오기
path = 'C:/Users/admin/Desktop/미니 프로젝트 1/데이터 통합용/전처리 필요/주거 관련/지역별_용도별_건물_수.xls'
df = pd.read_excel(path, header = 2)
df
```

```python
df_new = df
```

```python
# 안 쓰는 행 제거
df_new = df_new.drop([1, 20, 21, 22, 23]).reset_index(drop = True)
```

```python
# 컬럼명 변경
df_new = df_new.rename(columns = {'Unnamed: 0' : '지역'})
```

```python
# 확인
df_new
```

```python
# 안 쓰는 행 버리기
df_tmp = df_new.drop([0, 1]).reset_index(drop = True)
df_tmp
```

```python
# 연도만 추출하기
year_list = list()

for i in range(1, len(df_tmp.columns)) :
    if 'Unnamed: ' not in df_tmp.columns[i] :
        year_list.append(df_tmp.columns[i])
```

```python
# 확인
year_list
```

```python
# 연도별로 추출해서 빈 리스트에 넣기
df_list = list()

# 연도 리스트에서 연도 뽑아내는 용도
y_index = 0

# 6개 컬럼이 1개 연도임
for i in range(1, len(df_tmp.columns), 6) :
    # 리스트에 넣을 임시 테이터 프레임, for문 시작할 때마다 초기화 됨
    temp_df = pd.DataFrame()
    
    # 연도별로 뽑아서 임시 데이터 프레임에 넣음
    temp_df['지역'] = df_tmp['지역']
    temp_df['{} - 합계'.format(year_list[y_index])] = df_tmp[df_tmp.columns[i]]
    temp_df['{} - 주거용'.format(year_list[y_index])] = df_tmp[df_tmp.columns[i + 1]]
    temp_df['{} - 상업용'.format(year_list[y_index])] = df_tmp[df_tmp.columns[i + 2]]
    temp_df['{} - 공업용'.format(year_list[y_index])] = df_tmp[df_tmp.columns[i + 3]]
    temp_df['{} - 교육_사회용'.format(year_list[y_index])] = df_tmp[df_tmp.columns[i + 4]]
    temp_df['{} - 기타'.format(year_list[y_index])] = df_tmp[df_tmp.columns[i + 5]]
    
    # 임시 데이터 프레임을 위에서 만든 리스트에 넣음
    df_list.append(temp_df)
    
    # 연도 리스트 인덱싱용 변수 +1
    y_index += 1
```

```python
# 확인
df_list
```
