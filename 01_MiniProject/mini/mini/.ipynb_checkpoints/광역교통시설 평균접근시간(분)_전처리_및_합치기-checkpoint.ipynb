```python
import pandas as pd
```

```python
# 데이터 불러오기
path = 'C:/Users/admin/Desktop/미니 프로젝트 1/접근성 관련/2019년 광역교통시설 평균접근시간.csv'
df = pd.read_csv(path, header = 2)
```

```python
# 시간대별이라는 컬럼 제거
df_new = pd.DataFrame()
df = df.drop(['시간대별(1)'], axis = 1)
```

```python
# 확인
df
```

```python
# 사용할 컬럼 리스트
df.columns = (['지역','승용차_버스터미널','대중교통/도보_버스터미널',
               '승용차_철도역','대중교통/도보_철도역',
               '승용차_공항','대중교통/도보_공항'])
```

```python
# 지역별로 묶어서 더하기
df = df.groupby(['지역']).sum()
```

```python
# 전국이라는 행은 필요없음
df = df.drop(['전국'])
```

```python
# 새 데이터 프레임 만들고 세 컬럼에 값 넣음
df_new = pd.DataFrame()
df_new = df
df_new['버스터미널'] = round((df['승용차_버스터미널'] + df['대중교통/도보_버스터미널']) / 2, 2)
df_new['철도역'] = round((df['승용차_철도역'] + df['대중교통/도보_철도역']) / 2, 2)
df_new['공항'] = round((df['승용차_공항'] + df['대중교통/도보_공항']) / 2, 2)
```

```python
# 안 쓰는 컬럼 없애기
df_new = df_new.drop(['승용차_버스터미널', '대중교통/도보_버스터미널',
                      '승용차_철도역', '대중교통/도보_철도역',
                      '승용차_공항', '대중교통/도보_공항'], axis = 1)
```

```python
# 확인하기
df_new
```

```python
# 저장
df_new.to_csv('C:/Users/admin/Desktop/미니 프로젝트 1/데이터 통합용/df_광역교통시설 평균접근시간(분).csv')
```
