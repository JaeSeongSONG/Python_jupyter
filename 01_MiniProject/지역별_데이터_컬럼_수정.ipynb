```python
import pandas as pd
```

```python
df = pd.read_csv('./지역별_데이터.csv')
```

```python
df.columns
```

```python
df_new = pd.DataFrame()
```

```python
# 지역 컬럼 추가
df_new['지역'] = df['지역']
df_new
```

```python
# 컬럼 통합
df_new['고령화비율(%)'] = df['지역별 고령화 비율(%)']
df_new['경제활동인구(천명)'] = df['경제활동인구 (천명)']
df_new['청년인구'] = df['청년인구']
df_new['강소기업수'] = df['지역별 강소기업 갯수']
df_new['선호_사업군의_사업체_수'] = df['선호_사업군의_사업체_수']
df_new['매매가_전세가_비율'] = df['매매가격_대비_전세가격_비율']
df_new['문화(축제건수+문화시설수)'] = df['마을축제 건 수'] + df['전통문화축제'] + df['관광축제'] +\
                                      df['산업축제'] + df['특수목적축제'] + df['문화기반시설 수'] +\
                                      df['공연사업 수'] + df['공연 횟수'] + df['생활문화시설 수'] +\
                                      df['생활문화센터'] + df['평생 학습관'] +\
                                      df['문화의집'] + df['생활영상시설'] + df['작은도서관']
df_new['편의시설_평균_접근시간(분)'] = round((df['버스터미널'] + df['철도역'] + df['공항'] + df['병·의원'] + df['종합병원'] +\
                                          df['대규모점포'] + df['전통시장'] + df['초등학교'] + df['중학교'] + df['고등학교'] +\
                                          df['공공의료시설']) / 11, 2)
df_new['교통문화점수'] = round(df['지수(점)'], 2)
df_new['자연환경(휴양만족점수)'] = df['평균 휴양 만족 점수']
df_new['대기오염'] = df['총먼지(TSP)']
```

```python
df_new
```

```python
df_new.set_index(df_new['지역'], inplace = True)
```

```python
df_new.drop(['지역'],axis = 1 , inplace = True)
```

```python
# 동점 관측치 중에서 데이터 상에서 먼저 나타나는 관측치부터 순위 부여

df2['고령화비율순위'] = df_new['고령화비율(%)'].rank(method='first', ascending=False)
df2.sort_values('고령화비율순위').head(5)[['고령화비율순위']]
```
