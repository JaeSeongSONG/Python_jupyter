```python
# 라이브러리
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns

plt.rc('font', family = 'Malgun Gothic')
```

```python
# 파일 불러오기
path = 'C:/Users/admin/Desktop/미니 프로젝트 1/인구 관련/10년간 지역별 고령인구 추세.csv'
df = pd.read_csv(path, header = 1)
df
```

```python
# 사용하지 않은 행 제거하기
# 성별로 나눠져있는 행들 제거하고 합계만 사용함
for i in range(0, len(df)) :
    if '남자' in df['항목'][i] :
        df = df.drop([i])
    elif '여자' in df['항목'][i] :
        df = df.drop([i])
```

```python
# 전국 행 제거하기
df = df.drop([0])

# 인덱스 번호 재배열
df = df.reset_index(drop = True)
df
```

```python
# 항목 컬럼 제거
df = df.drop(['항목'], axis = 1)
df
```

```python
# '행정구역(동읍면)별 '이라는 컬럼명이 너무 길어서 변경
df = df.rename(columns = {'행정구역(동읍면)별' : '지역'})
df
```

```python
# 컬럼을 연도별로 나누기
div_list = list()

for i in range(1, 81, 8) :
    
    # 1열부터 81열까지 8단위씩 끊어서 출력  e.g.) i = 1, 9, 17, ... , 81
    
    tmp_list = list()
    
    for j in range(8) :
        
        tmp_list.append(df.columns[i])
        i += 1
        
        # 이중 for문을 통해 tmp_list에 df.columns[i]를 append하고, 
        # i + 1을 통해 range(8) 동안 1, 2, 3, 4, 5, 6, 7, 8의 열을 append
        
    div_list.append(tmp_list)
    
    # div_list에 tmp_list를 append
    # append 된 값은 한 set의 값을 가짐
```

```python
# 새 데이터 프레임
df_sum = pd.DataFrame()
df_sum['지역'] = df['지역']

# df_sum에 '지역'이라는 column을 만들고 df['지역'] 데이터를 넣음

# 연도 리스트 : 2012년 ~ 2021년
years = list(range(2012, 2022))

# 임시 리스트
temp = list()

for i in range(10) :
    df_sum[years[i]] = 0
    
# df_sum에 연도 리스트에 있는 값들을 이름으로 한 column을 생성

for div in div_list :
    for i in range(17) :
        temp.append(df[div].iloc[i].sum())
        
# div_list에 저장되어 있던 set의 i행을 더해서 temp에 append
        
index = 0
for year in years :
    for i in range(17) : 
        df_sum[year].iloc[i] = temp[index]
        index += 1
        
# df_sum의 각 열에 set의 더한 값들을 넣음
```

```python
df_sum
```

```python
pt = df_sum.pivot_table(values = [2012, 2013, 2014, 2015, 2016, 
                                  2017, 2018, 2019, 2020, 2021],
                        columns = '지역')
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(pt, annot = True, fmt = '.0f', cmap = 'rocket_r')
plt.title('10년간 고령 인구 변화 추세')
plt.show()
```
