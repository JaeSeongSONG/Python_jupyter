### # 특정 수치 이상이 되는 셀을 강조

```python
def color_negative_values(val):
  color = 'red' if val > 20 else 'black'
  return 'color: %s' % color

df4.style.applymap(color_negative_values)
```

```python
def color_max(s):
    is_max = s == s.max()
    return ['background-color: lightblue' if v else '' for v in    is_max]

df4.style.apply(color_max, axis=0)
```

```python
df4.style.applymap(color_negative_values).apply(color_max)
```
