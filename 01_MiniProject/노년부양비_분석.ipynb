```python
file_path = "./노인부양비1.csv"
df1 = pd.read_csv(file_path, header = 1)
df1
```

```python
df1.head()
```

```python
df4 = df1.drop(columns=["0"])
```

```python
df4.info()
```

```python
isin_filter = df4['년도'].isin([2000,2005,2010,2015,2020,2025,2030,2035,2040,2045,2050,2060])
df_old5 = df4[isin_filter]
print(df_old5)
```

```python
df_old5
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df_old5["년도"], df_old5["노년부양비"])
```

```python
import seaborn as sns
```

```python
# 노년부양비 추세 그래프 그리기
plt.rcParams['font.family'] = 'Malgun Gothic' # 폰트설정
plt.rcParams['axes.unicode_minus'] = False    # 깨짐 방지 
ax = plt.subplots()
ax = sns.barplot(x = "년도", y = "노년부양비", data = df_old5)
plt.ylim([0, 100])
ax.set_title("미래의 노년부양비")
ax.set_xlabel("년도")
ax.set_ylabel("미래의 노년부양비")
plt.grid() 
plt.savefig('노년부양비') # 그래프 저장

# 노년부양비
# 노년부양비 : 생산가능인구(15~64세 인구) 100명이 부양해야 할 65세 인구의 수를 표시
# 노년부양비 = (65세 이상인구 / 15세~64세 인구) × 100
```
