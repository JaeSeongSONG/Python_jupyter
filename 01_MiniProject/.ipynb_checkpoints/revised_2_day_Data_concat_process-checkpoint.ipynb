```python
# pandas 라이브러리 호출

import pandas as pd
```

```python
# '마을축제까지.csv' 파일 df로 불러오기

df=pd.read_csv('./마을축제까지.csv')
df
```

```python
# 'df_문화기반시설수.csv' 파일 df2로 불러오기
# df2['지역'] 열 삭제

df2 = pd.read_csv('./데이터 통합용/df_문화기반시설수.csv')
df2 = df2.drop('지역', axis = 1)
df2
```

```python
# ooncat 함수를 이용하여 df와 df2 데이터프레임을 열기준과 inner join의 형식으로 합침

df = pd.concat([df,df2],axis=1, join='inner')
df
```

```python
# 'df_문화예술 공연 건 수.csv' 파일 df2로 불러오기
# df2['지역'] 열 삭제

df2 = pd.read_csv('./데이터 통합용/df_문화예술 공연 건 수.csv')
df2 = df2.drop('지역', axis = 1)
df2
```
