#### - 시도의 경계가 나타나있는 json 파일을 불러오기
#### - 단, utf-8로 불러오는 것이 필수

```python
geo_path = './korea.json'
geo_str = json.load(open(geo_path, encoding = 'utf-8'))
```

#### - dust라는 리스트를 만들어 각 열의 이름을 넣음
#### - for 구문을 사용하여 dust에서 이름을 하나씩 빼낸 뒤
#### - df[i]에 하나씩 넣어 데이터 호출
#### - columns에는 지역이름값, df[i]값
#### - key_on에는 json파일의 feature.properties.CTP_KOR_NM 값을 넣음
#### - 단, 이 값은 위의 df.index값과 정확히 일치해야 함
#### - display(m) = plt.show() 와 같은 역할

```python
dust = ['총먼지(TSP)', '미세먼지(PM10)', '미세먼지(PM2.5)']

for i in dust:
    
    map = folium.Map(location = [36.8, 127.5], zoom_start = 7, width = 800, length = 300)
    
    map.choropleth(geo_data = geo_str,
                      data = df[i],
                      columns = [df.index, df[i]],
                      fill_color = 'Reds',
                      key_on = 'feature.properties.CTP_KOR_NM')
    display(map)
```
