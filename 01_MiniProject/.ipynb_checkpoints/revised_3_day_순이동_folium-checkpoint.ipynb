```python
# pandas와 folium 라이브러리 호출

import pandas as pd
import folium
```

#### - df를 [위도, 경도]로 순서대로 묶기

```python
location_data = df[['위도', '경도']].values.tolist()
location_data
```

```python
# folium.Map 을 이용하여 지도 그리기
# year의 값이 0보다 작으면 fill_color를 Red로, 0보다 크면 fill_color를 Blue 로 생성하여 지도에 add

m = folium.Map(location = location_data[6],
               zoom_start=6.5, 
               width=750, 
               height=700,
#                tiles='cartodbpositron'
              )

for i in range(17):
    
    year = int(df['2012 년'][i]) #  str 타입으로 출력하기 때문에 int타입으로 변경
    
    if year < 0:
        folium.CircleMarker(location_data[i], 
                        radius = (-1 * year) / 500, # radius = 반지름
                        color = None,
                       fill_color = 'Red').add_to(m)
        
    else:
        folium.CircleMarker(location_data[i], 
                        radius = year / 500,
                        color = None,
                       fill_color = 'Blue').add_to(m)

m
```

```python
df1 = pd.read_csv('./전국_순이동인구.csv')
year_list = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]


for i in year_list:
    
    m = folium.Map(location = location_data[6],
               zoom_start=6.5, 
               width=750, 
               height=700)
    
    Y = f'{i} 년'
    
    for j in range(17):
        
        year = int(df1[Y][j])
    
        if year < 0:
            folium.CircleMarker(location_data[j], 
                            radius = (-1 * year) / 500,
                            color = None,
                           fill_color = 'Red').add_to(m)

        else:
            folium.CircleMarker(location_data[j], 
                            radius = year / 500,
                            color = None,
                           fill_color = 'Blue').add_to(m)

    display(m) # plt.show와 같은 역할
```
